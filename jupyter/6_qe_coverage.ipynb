```python
%matplotlib widget
import scipp as sc
from trex.instrument import Instrument
from trex.components.source import Source
import plopp as pp
import matplotlib.pylab as plt

central_wavelength = 1.65 * sc.Unit("Å")
mode = "High Resolution"  # Chopper mode
rrm = 16
T_OFFSET = sc.scalar(1.7, unit="ms")

trex = Instrument(central_wavelength, rrm, mode, T_OFFSET)
trex.source = Source(
    facility="ess",
    neutrons=10_000_000,
)
res = trex.model.run()
```

```python
ei_ef_ratio = 0.0
# trex.unwrap_frame(res, ei_ef_ratio=0.4)
bounds = trex.estimate_qe_coverage(res, ei_ef_ratio)

fig, ax = plt.subplots()
p = pp.plot(bounds, ax=ax, markersize=1)
ax.set_title(str(trex) + f"\nei/ef = {ei_ef_ratio}", fontsize=8)
ax.set_ylim([-90, 90])
ax.grid(alpha=0.6)
fig.tight_layout()
```
