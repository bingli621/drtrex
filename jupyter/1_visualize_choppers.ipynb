```python
import scipp as sc
from trex.instrument import Instrument


T_OFFSET = sc.scalar(1.7, unit="ms")
central_wavelength = sc.scalar(1, unit="Å")

trex = Instrument(
    wavelength=central_wavelength, rrm=8, mode="High Flux", t_offset=T_OFFSET
)
print(trex)
```
